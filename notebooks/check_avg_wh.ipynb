## Resizing Images 

```python
dst = r"F:\Model\data\Weapons\weaponsbg\resized"
src = r"F:\Model\data\Weapons\weaponsbg\renamed"
```

```python
c = 0
for img_res in os.listdir(src):
    try:
        img = cv.imread(os.path.join(src,img_res))
        h,w,_ = img.shape
        if h < 500 and w < 500:
            c+=1
            os.remove(os.path.join(src,img_res))
            # shutil.move(os.path.join(src,img_res) ,r"F:\Model\data\fire&smoke\main\high_quality_train\images")
    except:
        print('non-type')

print(f" images which it's size over-equal than 600x600 are : {c} ")
```

```python
for img_res in os.listdir(src):
    try:
        img = cv.imread(os.path.join(src,img_res))
        
        resized = cv.resize(img, (640,640))
        
        cv.imwrite(os.path.join(dst,img_res) , resized)
    except:
        print('non-type')
```
